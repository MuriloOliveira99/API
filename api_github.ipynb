```python
import requests
import pandas as pd
```

```python
lista_usuario = []
lista_dados = []
while True:
  usuario = input('Digite seu usuário: ')

  if usuario == '':
    break
  lista_usuario.append(usuario)

  for usuario in lista_usuario:
    url = f'https://api.github.com/users/{usuario}'
    resposta = requests.get(url)
    usuario_info = resposta.json()

  if resposta.status_code == 200:
    # print(f"id: {usuario_info['id']}")
    # print(f"tipo: {usuario_info['type']}")
    # print(f"nome: {usuario_info['name']}")
    # print(f"email: {usuario_info['email']}")
    # print(f"blog: {usuario_info['blog']}")
    # print(f"bio: {usuario_info['bio']}")
    # print(f"repos_url: {usuario_info['repos_url']}")
    # print(f"foto: {usuario_info['avatar_url']}")

    dict_git = {"id": usuario_info['id'],
                  "tipo": usuario_info['type'],
                  "nome": usuario_info['name'],
                  "email": usuario_info['email'],
                  "blog": usuario_info['blog'],
                  "bio": usuario_info['bio'],
                  "repos_url": usuario_info['repos_url'],
                  "foto": usuario_info['avatar_url']
                  }
    lista_dados.append(dict_git)
  else:
    print('Usuário inexistente!')

df = pd.DataFrame(lista_dados) #['id', 'tipo', 'nome', 'email', 'blog', 'bio', 'repos_url', 'foto']
df.head()
```
